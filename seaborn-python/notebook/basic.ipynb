```python
import numpy as np
import pandas as pd
import seaborn as sns
```

# bar

```python
df = pd.DataFrame(
    data = {
        'type' : ['A', 'A', 'B', 'B', 'A', 'C', 'C'],
        'sub_type' : ['a', 'b', 'a', 'b', 'a', 'b', 'a'],
        'value' : [1, 2, 3, 4, 5, 6, 7]
    }
)
df
```

```python
sns.barplot(x='value', y='type', data=df, orient='h')
```

```python
sns.barplot(x='value', y='type', hue='sub_type', data=df, orient='h')
```

# histogram and continuous density

```python
arr_0 = np.random.normal(0, 1, size=200)
arr_1 = np.random.normal(10, 2, size=200)
ser = pd.Series(np.concatenate([arr_0, arr_1]))
sns.distplot(ser, bins=100, color='k')
```

```python
df = pd.DataFrame(
    data = {
        'x_val' : np.random.normal(0, 1, size=200),
        'y_val' : np.random.normal(10, 2, size=200)
    }
)
sns.regplot('x_val', 'y_val', data=df)
```

# pairs plot

```python
df = pd.DataFrame(
    data = {
        'val_0' : np.random.normal(0, 1, size=200),
        'val_1' : np.random.normal(10, 2, size=200),
        'val_2' : np.random.normal(0, 1, size=200),
        'val_3' : np.random.normal(10, 2, size=200)
    }
)
sns.pairplot(df, diag_kind='kde', plot_kws={'alpha': 0.2})
```

# factor plot

```python
df = pd.DataFrame(
    data = {
        'type' : ['A', 'A', 'B', 'B', 'A', 'C', 'C'],
        'sub_type' : ['a', 'b', 'a', 'b', 'a', 'b', 'a'],
        'value' : [1, 2, 3, 4, 5, 6, 7],
        'col' : ['x', 'x', 'x', 'y', 'y', 'y', 'y'],
        'row' : ['X', 'Y', 'X', 'Y', 'X', 'Y', 'X']
    }
)
sns.factorplot(x='value', y='type', hue='sub_type', col='col', data=df, kind='bar')
```

```python
sns.factorplot(x='value', y='type', hue='sub_type', col='col', row='row', data=df, kind='bar')
```

```python
sns.factorplot(x='value', y='type', kind='box', data=df)
```
