```python
import matplotlib.pyplot as plt
import matplotlib.animation

%matplotlib inline

plt.rcParams["animation.html"] = "jshtml"
```

```python
x = range(0, 10)
y = range(-5, 5)
fig, ax = plt.subplots()
line, = ax.plot([0, 10],[-5, 5])

animate = lambda i: line.set_data(x[:i], y[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(x) + 1)
ani
```
