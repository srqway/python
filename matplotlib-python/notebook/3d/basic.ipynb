```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

# plane

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
x, y = np.meshgrid(range(3), range(3))
z = x + y
ax.plot_surface(X = x, Y = y, Z = z)
plt.show()
```

# line

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(xs=[1, 2], ys=[3, 4],zs=[5, 6])
plt.show()
```

# array

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
x = np.array([0])
y = np.array([0])
z = np.array([0])
dx = np.array([1])
dy = np.array([2])
dz = np.array([3])
ax.quiver(x, y, z, dx, dy, dz)
ax.set_xlim(-1,4)
ax.set_ylim(-1,4)
ax.set_zlim(-1,4)
plt.show()
```
