# csv reader

```python
with open('/home/hsiehpinghan/git/python/pandas-python/data/csv/csvFileWithDoubleQuote.csv') as csvFileWithDoubleQuote:
    reader = csv.reader(csvFileWithDoubleQuote)
    for line in reader:
        print(line)
```

```python
class csv_reader_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ','
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL

with open('/home/hsiehpinghan/git/python/pandas-python/data/csv/csvFileWithDoubleQuote.csv') as csvFileWithDoubleQuote:
    reader = csv.reader(csvFileWithDoubleQuote, dialect=csv_reader_dialect)
    for line in reader:
        print(line)
```

# csv writer

```python
class csv_writer_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = '\t'
    quotechar = '"'
    quoting = csv.QUOTE_NONNUMERIC

with open('/tmp/csv_writer.csv', 'w') as f:
    writer = csv.writer(f, dialect=csv_writer_dialect)
    writer.writerow(('one', 'two', 'three'))
    writer.writerow((1, 2, 3))
    writer.writerow((4, 5, 6))
    writer.writerow((7, 8, 9))
    
%cat /tmp/csv_writer.csv
```
