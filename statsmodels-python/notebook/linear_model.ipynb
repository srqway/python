```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

# OLS (ordinary least squares)

```python
X = np.c_[
    [20, 30, 10, 40, 50],
    [32, 234, 23, 23, 42523]
]
X_model = sm.add_constant(X)
y = [10, 20, 30, 40, 50]
model = sm.OLS(y, X_model)
result = model.fit()
result.summary()
```

```python
data = pd.DataFrame(X, columns=['x_1', 'x_2'])
data['y'] = y
data[:5]
```

```python
result = smf.ols('y ~ x_1 + x_2', data=data).fit()
result.summary()
```
